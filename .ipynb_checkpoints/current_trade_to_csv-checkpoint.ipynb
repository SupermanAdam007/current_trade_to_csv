```python
import pandas as pd, glob, os, matplotlib.pyplot as plt, datetime.datetime as datetime
```

```python
def preprocess(df):
    df.drop_duplicates(inplace=True)
    df.drop('Shit', axis=1, inplace=True, errors='ignore')
    df.drop('ID', axis=1, inplace=True, errors='ignore')
    df = df[df.Market.notnull()]
    return df
```

```python
def parse_to_csv(df, fname):
    df.to_csv(fname, header=None, index=None, sep=';')
```

```python
os.chdir(os.getcwd())

for fname in glob.glob("trade*.txt"):
    print(fname)
    #'{SubscriptionId}~{ExchangeName}~{CurrencySymbol}~{CurrencySymbol}~{Flag}~{TradeId}~{TimeStamp}~{Quantity}~{Price}~{Total}'
    #Flag = 1: buy, Flag = 2: sell, Flag = 4: unknown 
    df_trade = pd.read_csv(fname, sep="~", header = None, names = ['ID', 'Market', 'From', 'To',
                                                                  'Flag', 'TradeId', 'Timestamp', 
                                                                  'Quantity', 'Price', 'Total', 'Shit'])
    df_trade = preprocess(df_trade)
    parse_to_csv(df_trade, fname.split('.')[0] + '.csv')

    
for fname in glob.glob("current*.txt"):
    print(fname)
    #'{Type}~{ExchangeName}~{FromCurrency}~{ToCurrency}~{Flag}~{Price}~{LastUpdate}~{LastVolume}~{LastVolumeTo}~{LastTradeId}~{Volume24h}~{Volume24hTo}~{MaskInt}'
    #Flag = 1: price up, Flag = 2: price down, Flag = 4: price unchanged
    df_current = pd.read_csv(fname, sep="~", header=None, names=['ID', 'Market', 'From', 'To',
                                                                'Flag', 'Price', 'Timestamp',
                                                                'LastVolume', 'LastVolumeTo',
                                                                'LastTradeId', 'Volume24h',
                                                                'Volume24hTo', 'Shit'])
    df_current = preprocess(df_current)
    parse_to_csv(df_current, fname.split('.')[0] + '.csv')
```

```python
timestamp_start = df_current['Timestamp'].iloc[0]
timestamp_end = df_current['Timestamp'].iloc[-1]
print(timestamp_start)
print(timestamp_end)
print(timestamp_end - timestamp_start)

print(datetime.fromtimestamp(int(timestamp_start)).strftime('%Y-%m-%d %H:%M:%S')
```

```python
plt.figure(figsize=(15, 8), dpi = 80, facecolor='w', edgecolor='k')
plt.plot(df_current.Price)
plt.show()
```
